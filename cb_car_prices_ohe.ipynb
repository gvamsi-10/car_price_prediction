# car price prediction using linear regression

### importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Downloads/carprices.csv")
df
```

# one hot encoding using pandas

```python
merge = pd.get_dummies(df['Car Model'])
merge
```

```python
df = pd.concat([df,merge],axis=1)
df
```

```python
fdf = df.drop(['Car Model','Mercedez Benz C class'],axis=1)
fdf
```

```python
X = fdf.drop('Sell Price($)',axis=1)
X
```

```python
y=fdf['Sell Price($)']
y
```

### model_fitting

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```
